```python
#looking at the ou sequences of the some knots
def get_enum_ou_string (knot):

    thing = ''.join(['.o_' if sign(j)==1 else '_u.' for j in knot.gauss_code()[0]])
    thing = [i.split('__') for i in thing.split('..')]
    thing = [ j for i in thing for j in i]
    thing = [i.replace('.','').replace('_','') for i in thing]
    thing = [ i[0]+str(len(i)) for i in thing]
    thing = ''.join(thing)
    return thing

def get_short_ou_string (knot):

    thing = ''.join(['.o_' if sign(j)==1 else '_u.' for j in knot.gauss_code()[0]])
    thing = [i.split('__') for i in thing.split('..')]
    thing = [ j for i in thing for j in i]
    thing = [i.replace('.','').replace('_','') for i in thing]
    thing = [ i[0] for i in thing]
    thing = ''.join(thing)
    return thing

ou = pd2ou(Knots().from_table(10,1).pd_code())
for i in range(100):
    print(get_short_ou_string(Knot(ou2pd(ou))))
    ou = R12Reduce(Gamma1(ou))
```

```python
# old code that has been rewritten in the other file but might
# be useful later?
def pd2dow (pd):
    eos = 2*len(pd)+1
    dow = [ [None,None] for i in pd]
    orientation = Knot(pd).orientation()
    
    s2x_dict = {}
    for i in range(len(pd)):
        s2x_dict[pd[i][0]] = [j for j in pd 
                                  if pd[i][2] in j and pd[i] != j][0]
        if orientation[i] == 1:
            s2x_dict[pd[i][1]] = [j for j in pd 
                                  if pd[i][3] in j and pd[i] != j][0]
        else:
            s2x_dict[pd[i][3]] = [j for j in pd 
                                  if pd[i][1] in j and pd[i] != j][0]
    strand = 1
    curr_x = pd[0]
    if orientation[pd.index(curr_x)] == 1:
        curr_pos = curr_x[1]
    else:
        curr_pos = curr_x[3]
    while strand < eos:
        if curr_pos == curr_x[0]:
            dow[pd.index(curr_x)][1] = strand
            curr_x, curr_pos = s2x_dict[curr_pos], curr_x[2]
        elif orientation[pd.index(curr_x)] == 1:
            dow[pd.index(curr_x)][0] = strand
            curr_x, curr_pos = s2x_dict[curr_pos], curr_x[3]
        else:
            dow[pd.index(curr_x)][0] = strand
            curr_x, curr_pos = s2x_dict[curr_pos], curr_x[1]
        strand += 1
    
    dow = [ tuple(x) for x in dow]
    dow = dict(zip(dow,orientation))
    dow[(eos,)] = None
    return dow
```
