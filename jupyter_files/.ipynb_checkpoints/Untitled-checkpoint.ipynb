Since Python does not have the same symbolic functionality as Mathematica, the representation of a tangle has to be adapted. Here we use a dict structure with key value pairs of the form (i,j):k, where i is the over strand, j is the understrand, and k is the crossing (either 1 or -1). For consistency EOSs (end of strands) are represented the same way except j is empty, and k is None.

As in the article, the i and j of a crossing need are preferably positive integers, so we now proceed with the Tidy function.

```python
def Tidy (tangle):
  ijs, ks = tangle.keys(), tangle.values()
  sort_ijs = sorted(list(set(sum(ijs,()))))
  new_ijs = [tuple(sort_ijs.index(s)+1 for s in ij) for ij in ijs]
  return dict(zip(new_ijs, ks))
```

We then try to Tidy a virtual diagram. We use the ame example as in the article, however we notice that the output is different from what the Tidy function in the article does, and that is due to an issue with the example.

Mathematica sorts the entries differently depending on their "type", the symbols pi and e used in the Mathematica code are just symbols they do not represent the numerical values we associate them with, and are sorted by their own order, instead of the numerical value that they should represent. In short, the example in the article is bad because the code assumes that the user inputs real numbers, and not symbols. If one replaces pi and e in the article's example to 3.141 and 2.718, respectively, the code outputs the same result as here.

Now the R12Reduce and R12Reduce1 functions

```python
def R12Reduce1 (tangle):
  
  xs, ends = {}, {}
  for ijs, ks in tangle.items():
    if ks==None: ends[ijs]=ks
    else: xs[ijs]=ks

  if len(xs) == 0: return tangle
  
  R2s = [[(x,y),(x+1,y+1)] for (x,y),s in xs.items() if ((x+1,y+1),-s) in xs.items()]
  if  len(R2s)>0:
    for item in R2s[0]: xs.pop(item)
    xs.update(ends)
    return Tidy(xs)
  
  R2s = [[(x,y),(x+1,y-1)] for (x,y),s in xs.items() if ((x+1,y-1),-s) in xs.items()]
  if len(R2s) >0:
    for item in R2s[0]: xs.pop(item)
    xs.update(ends)
    return Tidy(xs)
  
  xs = { x:y for x,y in xs.items() if abs(x[0]-x[1]) != 1}
  xs.update(ends)
  return Tidy(xs)

def R12Reduce (tangle):
  old = tangle
  new = R12Reduce1(tangle)
  while old != new:
    old = new
    new = R12Reduce1(new)    
  return old
```

We move to the next function which is the glide function Gamma1.

```python
def Gamma1 (tangle):
  
  new_tangle = tangle.copy()
  xs, ends = {}, {}
  for ijs, ks in new_tangle.items():
    if ks==None:
        ends[ijs]=ks
    else:
        xs[ijs]=ks
        
  ijs, ks = zip(*xs.items())
  os = [ i[0] - 1 for i in ijs]
  uos =  sorted([ j[1] for j in ijs if j[1] in os])
  
  if len(uos) == 0: return new_tangle
  else:
    j1 = uos[0]
    i2 = j1 + 1
    
    jss = list(zip(*ijs))[1]
    iss = list(zip(*ijs))[0]
    
    
    idx_j1 = list(zip(*ijs))[1].index(j1)
    idx_i2 = list(zip(*ijs))[0].index(i2)
        
    i1 = ijs[idx_j1][0]
    j2 = ijs[idx_i2][1]
    
    s1 = ks[idx_j1]
    s2 = ks[idx_i2]
    
    new_tangle[j1, j2] = s2
    new_tangle[i1, i2] = s1
    new_tangle[(i1 - s1/3, j2 + s2/3)] = s1 * s2
    new_tangle[(i1 + s1/3, j2 - s2/3)] = - s1 * s2
    
    if ijs[idx_j1]==ijs[idx_i2]:
        new_tangle.pop(ijs[idx_j1])
    else:
        new_tangle.pop(ijs[idx_j1])
        new_tangle.pop(ijs[idx_i2])
        
  return Tidy(new_tangle)
```

One issue with the above function, and also the code in the article, is that the implementation does not properly handle tangles with crossings of the form (n+1,n):\pm1, we illustrate the issue below:

```python
issue = {(2,1):1,(3,):None}

Gamma1(issue)
```

The issue is that you get extra entries (2,2) and (5,5) which do not describe valid crossings. In the article this does not pose an issue, since after each iteration of Gamma1, there is an application of R12Reduce, so these troublesome crossings are removed by an R1 move. However, if one chooses not to use R12Reduce at some point, then the above issue should be kept in mind.

Since the knot diagram representation used in the article is non-standard, we first begin by making a pair of functions that convert the dowker-like notation from the article to the planar diagram presentation as outlined in the documentation of the Sagemath package. 

One main difference between the packages of Sagemath and Mathematica is that the strands for each crossing are counted clockwise and counterclockwise, respectively.

```python
# currently this code supports only tangles of 1 component
# because I haven't figured out how to do the general case yet
def ou2pd (ou):
    
    eos_idx = list(ou.values()).index(None)
    last_s = list(ou.keys())[eos_idx][0]-1
    
    pd = []
    for x in ou.keys():
        if ou[x] == None: pass
        elif ou[x] == -1:
            pd += [[x[1], x[0]%last_s+1, x[1]%last_s+1, x[0]]]
        else:
            pd += [[x[1], x[0], x[1]%last_s+1, x[0]%last_s+1]]
    return pd

def pd2ou (pd):
    eos = 2*len(pd)+1
    ou = [ [None,None] for i in pd]
    orientation = Knot(pd).orientation()
    
    s2x_dict = {}
    for i in range(len(pd)):
        s2x_dict[pd[i][0]] = [j for j in pd 
                                  if pd[i][2] in j and pd[i] != j][0]
        if orientation[i] == 1:
            s2x_dict[pd[i][1]] = [j for j in pd 
                                  if pd[i][3] in j and pd[i] != j][0]
        else:
            s2x_dict[pd[i][3]] = [j for j in pd 
                                  if pd[i][1] in j and pd[i] != j][0]
    strand = 1
    curr_x = pd[0]
    if orientation[pd.index(curr_x)] == 1:
        curr_pos = curr_x[1]
    else:
        curr_pos = curr_x[3]
    while strand < eos:
        if curr_pos == curr_x[0]:
            ou[pd.index(curr_x)][1] = strand
            curr_x, curr_pos = s2x_dict[curr_pos], curr_x[2]
        elif orientation[pd.index(curr_x)] == 1:
            ou[pd.index(curr_x)][0] = strand
            curr_x, curr_pos = s2x_dict[curr_pos], curr_x[3]
        else:
            ou[pd.index(curr_x)][0] = strand
            curr_x, curr_pos = s2x_dict[curr_pos], curr_x[1]
        strand += 1
    
    ou = [ tuple(x) for x in ou]
    ou = dict(zip(ou,orientation))
    ou[(eos,)] = None
    return ou
```

An issue present in the above two functions is that firstly they only work for knots and not for links, secondly, the implementations make a few assumptions. 

The PD code represents a knot or a closed tangle, meanwhile the Dowker-like notation represents an open tangle. This is an issue since converting from a PD code to the Dowker-like notation requires us to cut the knot somewhere. This makes the translation not unique, and possibly provides some unexpected effects to the results of the algorithm. In other words, it is likely that the gliding algorithm produces different results for depending on where we cut the knot.

The pd2ou function works like this: start at a crossing, pick a number representing the start of a strand, then follow the number through each crossing while enumerating each strand you visit. The convention used in the pd2ou function above is to start at the first crossing of the PD code, and to begin at the start of the overstrand. We do this because we want to put the knot into OU form, might as well orient the first crossing correctly.

Finally, the ou2pd function is not the inverse of pd2ou, since you do not get the same PD code back as demonstrated below with the trefoil knot. This is due to the fact that the connections in the PD codes from the Rolfsen table are named at random, not following any particular orientation. This can be remedied by writing another function to rename the connections in the PD code, and possibly some minor changes to the above two functions but that seems extra for now.

```python
def get_knot_pics (iter_func,knot,name, composite =False):
    
    dir_cont = [ i for i in os.listdir() if name+'_plots' in i]
    if len(dir_cont) == 0:
        dir_name = './'+name+'_plots_0/'
    else:
        latest = max(dir_cont)
        new_numb = str(int(latest.split('_')[-1])+1)
        dir_name = './'+name+'_plots_'+new_numb+'/'
    os.mkdir(dir_name)
    
    ou = pd2ou(knot.pd_code())
    for i in range(100):
        try:
            the_plot = knot.plot()
            the_plot.save(dir_name+name+str(i)+'.png')
            ou = iter_func(ou)
            knot = Knot(ou2pd(ou))
        except ValueError as ve:
            print('something went wrong, probably a crossing of type (2,1):1')
            return 'you stopped at iteration: '+str(i)
    
    new_size = (500,500)
    
    for i in range(100):
        bkgnd = Image.new('RGB', new_size, (255,255,255))

        f_name = dir_name+name+str(i)+'.png'
        knot_plot = Image.open(f_name)
        knot_plot.convert('RGBA')

        curr_size = knot_plot.size
        dw = new_size[0]-curr_size[0]
        dh = new_size[1]-curr_size[1]
        padding = (dw//2,dh//2, dw-dw//2, dh-dh//2)

        knot_plot = expand(knot_plot, padding)
        knot_plot = knot_plot.save(f_name)

    if composite == True:
        list_im = [ dir_name+name+str(i)+'.png' for i in range(100)]
        imgs = [Image.open(i) for i in list_im]

        imgs_h_comb = []
        for j in range(10):
            imgs_h_comb += [hstack( [asarray(i) for i in imgs[10*j:10*(j+1)]] )]

        imgs_v_comb = vstack( [i for i in imgs_h_comb])
        imgs_v_comb = Image.fromarray(imgs_v_comb).save(dir_name+name+'100arr.png')
```

We see that in the above few iteration, the knot diagram begins to accumulate around a spurious kink, specifically at the center of the spiral there is an R1 crossing. This crossing should not be there, since we always reduce the diagram modulo R1 and R2 moves. This crossing evades our sight because it is actually produced by connecting the end points of the tangle. Our functions do not see this extra connection, hence cannot reduce the crossing.

We propose an addition to the methods.

```python
def R12Reduce1 (tangle):
  
  xs, ends = {}, {}
  for ijs, ks in tangle.items():
    if ks==None: ends[ijs]=ks
    else: xs[ijs]=ks

  # we assume that there is only one end of strand (1-tangle)
  eos = list(ends.keys())[0][0]-1

  if len(xs) == 0: return tangle

  for n in range(eos+1):
    if (eos,n) in xs.keys() and (1,n+1) in xs.keys() and xs[(eos,n)]==-xs[(1,n+1)]:
      xs.pop((eos,n)); xs.pop((1,n+1))
    if (n,eos) in xs.keys() and (n+1,1) in xs.keys() and xs[(n,eos)]==-xs[(n+1,1)]:
      xs.pop((n,eos)); xs.pop((n+1,1))
  R2s = [[(x,y),(x+1,y+1)] for (x,y),s in xs.items() if ((x+1,y+1),-s) in xs.items()]
  if  len(R2s)>0:
    for item in R2s[0]: xs.pop(item)
    xs.update(ends)
    return Tidy(xs)
  
  for n in range(eos+1):
    if (eos,n) in xs.keys() and (1,n-1) in xs.keys() and xs[(eos,n)]==-xs[(1,n-1)]:
      xs.pop((eos,n)); xs.pop((1,n-1))
    if (n,1) in xs.keys() and (n+1,eos) in xs.keys() and xs[(n,1)]==-xs[(n+1,eos)]:
      xs.pop((n,1)); xs.pop((n+1,eos))
  R2s = [[(x,y),(x+1,y-1)] for (x,y),s in xs.items() if ((x+1,y-1),-s) in xs.items()]
  if len(R2s) >0:
    for item in R2s[0]: xs.pop(item)
    xs.update(ends)
    return Tidy(xs)
  
  
  xs = { x:y for x,y in xs.items() if abs(x[0]-x[1]) != 1}
  if (1,eos) in xs.keys():
    xs.pop((1,eos))
  elif (eos,1) in xs.keys():
    xs.pop((eos,1))
    
  xs.update(ends)
  return Tidy(xs)

def R12Reduce (tangle):
  old = tangle
  new = R12Reduce1(tangle)
  while old != new:
    old = new
    new = R12Reduce1(new)    
  return old
```

If we run the loop now, we see we do not get that R1 crossing anymore. Further we see that with this diagram, we no longer have accumulation, instead we have a alternating sequence of length 2. This might be a recurring theme for the rest of our work.
